### 2 Euler's formula* $ e^{j\varphi}=cos\varphi+jsin\varphi$* $ cos\varphi = \frac{e^{j\varphi}+e^{-j\varphi}}{2}$* $ sin\varphi = \frac{e^{j\varphi}-e^{-j\varphi}}{2j}$

### 3 Complex sinewave$ \bar{x}[n]=Ae^{j(\omega nT+\varphi)} = A cos(\omega nT + \varphi) + j A sin(\omega nT + \varphi)$ $x[n]=Acos(\omega nT + \varphi) =  \frac{e^{j(\omega nT + \varphi)}+e^{-j(\omega nT + \varphi)}}{2}$$=\frac{1}{2}Xe^{j(\omega nT)} + \frac{1}{2}X^*e^{-j(\omega nT)}$

```python
real = A * cos(2*np.pi*f*t + phi)imag = A * sin(2*np.pi*f*t + phi)plot(t,real)plot(t,imag)legend(["real","imag"])
```
