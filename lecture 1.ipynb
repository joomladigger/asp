### 2 Euler's formula* $ e^{j\varphi}=cos\varphi+jsin\varphi$* $ cos\varphi = \frac{e^{j\varphi}+e^{-j\varphi}}{2}$* $ sin\varphi = \frac{e^{j\varphi}-e^{-j\varphi}}{2j}$

### 3 Complex sinewave$ \bar{x}[n]=Ae^{j(\omega nT+\varphi)} = A cos(\omega nT + \varphi) + j A sin(\omega nT + \varphi)$ $x[n]=Acos(\omega nT + \varphi) =  \frac{e^{j(\omega nT + \varphi)}+e^{-j(\omega nT + \varphi)}}{2}$$=\frac{1}{2}Xe^{j(\omega nT)} + \frac{1}{2}X^*e^{-j(\omega nT)}$

```python
real = A * cos(2*np.pi*f*t + phi)imag = A * sin(2*np.pi*f*t + phi)plot(t,real)plot(t,imag)legend(["real","imag"])
```

```python
## plot complex-sinewaveimport matplotlib.pyplot as pltimport numpy as npN = 500k = 3plt.figure(1,figsize(9.5,6))s = np.exp(1j*2*np.pi*k/N*np.arange(-N/2,N/2))
```

```python
plt.plot(np.arange(-N/2,N/2),np.real(s),'b',label='real')plt.plot(np.arange(-N/2,N/2),np.imag(s),'g',label='imag')plt.xlabel('time')plt.ylabel('amplitude')plt.axis([-N/2,N/2,-1,1])plt.legend()plt.tight_layout()
```

### 4 Convolution

```python
import matplotlib.pyplot as pltimport numpy as npfrom scipy.signal import hamming, triang, blackmanharristriangle = np.zeros (70)triangle[1:32] = triang (31)rectangle = np.zeros (70)rectangle[1:32] = np.ones (31)output = np.zeros (70)output = np.convolve(triangle, rectangle)plt.figure(1, figsize=(9.5, 5.5))plt.subplot(3, 1, 1)plt.plot(triangle, lw=2)plt.axis([0,69,-.1,1.1])plt.title ('x1 (triangle)')plt.subplot(3, 1, 2)plt.plot(rectangle, lw=2)plt.axis([0,69,-.1,1.1])plt.title ('x2 (rectangle)')plt.subplot(3, 1, 3)plt.plot(output, lw=2)plt.axis([0,69,-1,17])plt.title ('y = x1 * x2')
```
